# Working with `earthaccess` results the new way

**Purpose**: Simplify and enhance user interaction with results of `earthaccess.search_data` or `.search_datasets`. Currently, users need expertise with various tools to leverage results lists and cannot trace provenance (i.e. search parameters), motivating this change.

**Outcome**: User can more easily visualize, reproduce, convert between formats, and understand the results.

**Process**: Create a new `Results` class with **a single responsibility -- describing results, including how we got them**. Enable easy conversion to other formats (geodataframes, disk format e.g. `results.save(...)`/`Results.load()`, ...), accessing the provenance of the results (e.g. original query parameters, time executed, ...), easy visualization (`results.explore()`), ...?

For this demo, we've implemented basic provenance, conversion to geodataframe, and pretty display in Jupyter Notebooks.

## What's next

### `QueryParameters` interface

**Single responsibiliy: handles what you want to communicate, not how you communicate.**

```python
query = earthaccess.GranuleQuery(
    concept_id=...,
)
results = earthaccess.search(query)
```

Notice that we didn't say `search_data` or `search_datasets` (those are confusing names, right?), the query object entirely expresses what the user wants.

### Save and load `Results`!

**Enables reproducibility!**

```python
results = earthaccess.search(...)
results.save("/my-results.json")
```

...

```python
results = Results.load("/my-results.json")

# Reproduce:
earthaccess.download(results)

# You can also repeat:
earthaccess.download(earthaccess.search(results.query))
```
